# Que1.● The mean sales generated by each zone

```python
ZA= np.array(data['Zone - A'])
ZB= np.array(data['Zone - B'])
ZC= np.array(data['Zone - C'])
ZD= np.array(data['Zone - D'])
```

```python
A=np.mean(ZA)
print('mean of Zone A:',A)
B=np.mean(ZB)
print('mean of Zone B:',B)
C=np.mean(ZC)
print('mean of Zone C:',C)
D=np.mean(ZD)
print('mean of Zone D:',D)
```

# Que3. ● Check whether all the zones generate the same amount of sales.

```python
from scipy.stats import f_oneway
F,p = f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
print("The F-statistic is {0} \n p-value is {1}".format(F,p))
```

```python
if p<0.05:
    print("Reject null hypothesis, which indicates zones generate different amount of sales")
else:
    print("Accept null hypothesis, which indicates all the zones generate same amount of sales.") 
```
